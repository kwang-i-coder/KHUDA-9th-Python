```python
# @title KHUDA 9기 전용 자동 채점기
import base64

class KHUDA_Grader:
    def __init__(self):
        self.answers = {
            1: "SGVsbG8gV29ybGQh",
            2: "WzEwLCA0LCAyMSwgMiwgMV0=",
            3: "WzIzNjAsIDM3NzYsIDE0MTYsIDE4MTcyMF0=",
            4: "PA==",
            5: "WzksIDI1XQ==",
            6: "Mw==",
            7: "Mw==",
            8: "WzMsIDEsIDQsIDIsIDVd",
            9: "NDM3",
            10: "WzEsIDIsIDMsIDQsIDVd"
        }

    def check(self, task_id, user_func, *args):
        try:
            actual = user_func(*args)
            decoded = base64.b64decode(self.answers[task_id]).decode()
            
            # 리스트나 정수형으로 복원
            if decoded.startswith('['): expected = eval(decoded)
            elif decoded.isdigit(): expected = int(decoded)
            else: expected = decoded

            if actual == expected:
                print(f"Task {task_id}: 정답입니다!")
            else:
                print(f"Task {task_id}: 틀렸습니다. 결과값 '{actual}'을(를) 다시 확인하세요.")
        except Exception as e:
            print(f"Task {task_id}: 에러 발생! 코드를 완성했는지 확인해주세요. ({e})")

grader = KHUDA_Grader()
print("채점 준비 완료!")
```
