![Task6_1](https://raw.githubusercontent.com/kwang-i-coder/KHUDA-9th-Python/main/images/task6_1.png)

코딩 테스트를 풀다 보면 로직은 맞는데 **시간 초과**로 틀리는 경우가 있습니다. 그 주범은 의외로 `input()` 함수일 때가 많아요!

* **`input()`**: 하나씩 정중하게 요청해서 느림
* **`sys.stdin.readline`**: 버퍼를 사용해 통째로 읽어와서 훨씬 빠름

#### 주의할 점
`readline`은 줄 끝의 엔터(`\n`)까지 통째로 읽어옵니다. 그래서 반드시 뒤에 `.split()`이나 `.strip()`을 붙여서 불필요한 공백을 털어내야 합니다.

#### 실전 활용 예시
데이터가 100만 개쯤 들어오는 복잡한 문제에서는 반드시 아래와 같이 작성해야 합니다.

```python
import sys

for _ in range(1000000):
    a, b = map(int, sys.stdin.readline().split())
    print(a + b)

### 7. 입력이 더 이상 들어오지 않을 때까지 반복하는 구조를 익힙니다.
핵심 개념: try-except 구문을 이용한 EOF(End Of File)

try: 평상시 실행될 코드를 작성
except: 더 이상 읽을 데이터가 없어서 에러가 발생하면, 프로그램이 멈추지 않고 안전하게 종료됩니다.

![Task7](https://raw.githubusercontent.com/kwang-i-coder/KHUDA-9th-Python/main/images/task7.png)

---
어떤 문제는 "숫자를 10개 줄게"라고 알려주지만, 어떤 문제는 "언제 끝날지 안 알려줄 테니 다 읽으면 멈춰"라고 합니다. 이때 사용하는 개념이 EOF (End Of File)입니다.

#### `try-except` 구조
더 이상 읽을 데이터가 없는데 억지로 읽으려고 하면 파이썬은 `EOFError`를 냅니다. 이걸 미리 대비

* `try:`: 평소처럼 데이터를 계속 읽어
* `except EOFError:`: 데이터가 다 떨어졌네? => 그럼 에러 내지 말고 여기서 멈춰

> 실전 활용 예시
> ```py
> while True:
>     try:
>         a, b = map(int, input().split())
>         print(a + b)
>     except EOFError:
>         break # 입력이 끝나면 무한 루프 탈출
> ```
